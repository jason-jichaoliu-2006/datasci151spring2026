**Q1** Recode a numeric column

<font size = "3">

- Read in the file "movies.csv" from the `quiz_data` folder, saving it as a DataFrame.

- Recode the "year" variable to create a new variable called "time_period" with the following categories.

$\qquad$ ``` ["1941-1967","1968-2003","2004-onwards"] ```

- You must use the `pd.cut` function to recode this variable.

- Print the DataFrame to the screen to check your answers


````python
# your code here

# Read the CSV file
df = pd.read_csv("quiz_data/movies.csv")

# Define bins and labels
bins = [1940, 1967, 2003, float('inf')]
labels = ["1941-1967", "1968-2003", "2004-onwards"]

# Recode 'year' into 'time_period'
df["time_period"] = pd.cut(df["year"], bins=bins, labels=labels)

# Print the DataFrame to check
print(df)
````

**Q2** Aggregate and Sort

<font size = "3">

- Continue to use the DataFrame from the previous question.

- Do the following in a **single line of code** using chaining:

    - Group the data by "country"
    
    - Aggregate the following statistics: average runtime, maximum runtime, average IMDB score, minimum IMDB score

    - Sort by average IMDB score in **descending order**

- Print the resulting DataFrame to the screen

**Note:** A "single line of code" can continue on to the next line if you use a backslash or wrap the line within a pair of parentheses.

````python
# your code here
print(
    df.groupby("country")
      .agg(avg_runtime=("runtime(minutes)", "mean"),
           max_runtime=("runtime(minutes)", "max"),
           avg_imdb=("imdb_score", "mean"),
           min_imdb=("imdb_score", "min"))
      .sort_values(by="avg_imdb", ascending=False)
)
````

**Q4** Aggregate and query

<font size = "3">

- Read in the file "sprint_results.csv" from the `quiz_data` folder, saving it as a DataFrame.

- Do the following in a single line using chaining:

    - Extract the subset with constructorId equal to 51.

    - Aggregate the maximum number of laps and average number of laps for each driver in the subset.

    - Extract the subset of these drivers with maximum # of laps $\geq 20$. 


````python
# your code here

print(
    pd.read_csv("quiz_data/sprint_results.csv")
      .query("constructorId == 51")
      .groupby("driverId")
      .agg(max_laps=("laps", "max"),
           avg_laps=("laps", "mean"))
      .query("max_laps >= 20")
)
````

**Q5** Calculate race duration

<font size = "3">

- Read in the file "results_cleaned.csv" from the `quiz_data` folder, saving it as a DataFrame.

- The "milliseconds" column shows the race duration for each result measured in milliseconds. Create a new column called "race_duration_minutes" where we convert the race duration to minutes by dividing each millisecond value by 60,000 (which equals $6 \times 10^4$). Equivalently, you can multiply each millisecond value by $6 \times 10^{-4}$

- Group the data by **both** "raceId" and "constructorId" and calculate the average race duration in minutes for each group.

- Sort the data in descending order and then extract the raceId/constructorId pairs with the 5 slowest average times.


````python
# your code here

# Read data
df = pd.read_csv("quiz_data/results_cleaned.csv")

# Create new column (milliseconds → minutes)
df["race_duration_minutes"] = df["milliseconds"] / 60000

# Group and compute average
grouped = df.groupby(["raceId", "constructorId"])["race_duration_minutes"].mean()

# Sort in descending order
sorted_result = grouped.sort_values(ascending=False)

# Get top 5 slowest
result = sorted_result.head(5)

# Print result
print(result)
````

**Q6** Aggregate custom statistic

<font size = "3">

- Continue to use the DataFrame corresponding to the data in "results_cleaned.csv".

- When a driver finishes 1st, 2nd, or 3rd it is known as a "podium finish". Write a function that takes in a Pandas Series and returns the number of entries $\leq 3$.

- Group by "driverId" and then aggregate the following statistics: the average of the "points" column, and the number of podium finishes in the "positionOrder" column - use your custom function for this second statistic.

- Merge the aggregated data back into the original DataFrame.

````python
# your code here


# Load correct dataset
df = pd.read_csv("quiz_data/results_cleaned.csv")

# Custom function
def count_podium(series):
    return (series <= 3).sum()

# Aggregate
agg_df = df.groupby("driverId").agg(
    avg_points=("points", "mean"),
    podium_count=("positionOrder", count_podium)
).reset_index()

# Merge back
df = df.merge(agg_df, on="driverId", how="left")

print(df)
````
